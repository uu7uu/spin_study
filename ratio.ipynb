```python
import numpy as np
from astropy.io import fits
from matplotlib import pyplot as plt
import matplotlib.ticker
```

```python
def xx(data):
    i = []
    for n in range(len(data)):
        _i = data[n][0]
        i.append(float(_i))
    return i

def xee(data):  
    a = []
    for n in range(len(data)):
        _a = data[n][1]
        a.append(float(_a))
    return a


def yy(data):
    b = []
    for n in range(len(data)):
        _b = data[n][2]
        b.append(float(_b))
    return b


def yee(data):
    c = []
    for n in range(len(data)):
        _c = data[n][3]
        c.append(float(_c))
    return c

```

```python
data=[]
with open('7.txt','r') as f:
    for line in f:
        data.append(list(line.strip('\n').split(' ')))
print(data[0])
x7l=xx(data)
xe7l=xee(data)
y7l=yy(data)
ye7l=yee(data)
```
