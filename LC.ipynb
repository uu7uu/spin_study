```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("glcbin24.0h_regbg_v1.csv",names=['mjd','2','3','2_20','2e20','2_4','2e4','4_10','4e10','10_20','10e20','0'])
print(data)
```

```python
mjd = data["mjd"]
c2_20 = data["2_20"]
c4_20 = data["4_10"] + data["10_20"]
c2_4 = data["2_4"]
e2_4 = data["2e4"]
e2_20 = data["2e20"]
e4_20 = e2_20**2 +e2_4**2

hr = c4_20/c2_4
ehr = e4_20**2/c2_4**2 + e2_4**2*c4_20**2/c2_4**4
```

```python
plt.figure(figsize=(15,15))
ax = plt.subplot(211)
ax.set_xlim(58495,58635)
ax.set_ylim(-1,11)
plt.yticks(fontsize=24)
ax.set_ylabel("$\mathrm{photon \quad cm^{-2} \quad s^{-1}}$", fontsize=24)
plt.errorbar(mjd,c2_20,e2_20,linestyle="none",fmt="k",marker='o',markersize=3)
ax.axvline(x=58510,color="green",linewidth=2) #1
ax.axvline(x=58511,color="green",linewidth=2) #2
ax.axvline(x=58512,color="green",linewidth=2) #3
ax.axvline(x=58518,color="green",linewidth=2) #6
ax.axvline(x=58519,color="red",linewidth=2)
ax.axvline(x=58520,color="red",linewidth=2) #11
ax.axvline(x=58522,color="magenta",linewidth=2) #17
ax.axvline(x=58527,color="magenta",linewidth=2) #18
ax.axvline(x=58528,color="blue",linewidth=2) #19
ax.axvline(x=58529,color="blue",linewidth=2)
#ax.axvline(x=58533,color="blue",linewidth=2) #21
#ax.axvline(x=58534,color="blue",linewidth=2) #22

#plt.fill_between([58519,58529],-1,11,facecolor = 'salmon', alpha = 0.4)

bwith = 2
ax.spines['left'].set_linewidth(bwith)
ax.spines['right'].set_linewidth(bwith)
ax.spines['top'].set_linewidth(bwith)
ax.spines['bottom'].set_linewidth(bwith)

ax = plt.subplot(212)
ax.set_xlim(58495,58635)
ax.set_ylim(-0.2,2.8)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
ax.set_xlabel("MJD", fontsize=24)
ax.set_ylabel("Hardness  Ratio", fontsize=24)
plt.errorbar(mjd,hr,ehr,linestyle="none",fmt='k',marker='o',markersize=3)
ax.axvline(x=58510,color="green",linewidth=2) #1
ax.axvline(x=58511,color="green",linewidth=2) #2
ax.axvline(x=58512,color="green",linewidth=2) #3
ax.axvline(x=58518,color="green",linewidth=2) #6
ax.axvline(x=58519,color="red",linewidth=2) #7
ax.axvline(x=58520,color="red",linewidth=2) #11
ax.axvline(x=58522,color="magenta",linewidth=2) #17
ax.axvline(x=58527,color="magenta",linewidth=2) #18
ax.axvline(x=58528,color="blue",linewidth=2) #19
ax.axvline(x=58529,color="blue",linewidth=2) #8
#ax.axvline(x=58533,color="blue",linewidth=2) #21
#ax.axvline(x=58534,color="blue",linewidth=2) #22

#plt.fill_between([58519,58529],-0.2,2.8,facecolor = 'salmon', alpha = 0.4)

ax.spines['left'].set_linewidth(bwith)
ax.spines['right'].set_linewidth(bwith)
ax.spines['top'].set_linewidth(bwith)
ax.spines['bottom'].set_linewidth(bwith)
plt.subplots_adjust(hspace=0)
plt.savefig("HR_new.png", dpi=200, bbox_inches='tight')
```
