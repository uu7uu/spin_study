```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np
 
def fund(x, a, b):
    return a*x**b



tt1 = [0.617,0.647,0.780,0.755,0.739,0.733]
tt2 = [0.640,0.650,0.784,0.756,0.739,0.733]

F = [7.5952,7.8765,8.9044,7.7538,6.8927,6.5014]
F = [0.5372,0.6675,1.6593,1.4682,1.7131,1.7059]
#F = [1.7857,1.8808,2.3724,2.1462,2.0973,2.0393]

L = F
t1 = tt1
t2 = tt2 


plt.scatter(t1,L,color='k',marker='x')

popt, pcov = curve_fit(fund, t1, L, maxfev=5000)
# #popt数组中，三个值分别是待求参数a,b,c
y1 = [fund(i, popt[0],popt[1]) for i in t1]
yy1=sorted(y1)
t11=sorted(t1)
print(yy1)
plt.plot(t11,yy1,"r--")

bwith = 2 #边框宽度设置为2
ax = plt.gca()#获取边框
ax.spines['bottom'].set_linewidth(bwith)
ax.spines['left'].set_linewidth(bwith)
ax.spines['top'].set_linewidth(bwith)
ax.spines['right'].set_linewidth(bwith)
plt.xticks(fontsize=15)
plt.yticks(fontsize=18)
plt.xlabel(r"$k\mathrm{T_{in}(keV)}$",fontsize=24)
plt.ylabel(r"$\mathrm{Flux(10^{-7}\ erg\ cm^{-2}\ s^{-1})}$",fontsize=24)


#table4 画图
print(popt)
```
