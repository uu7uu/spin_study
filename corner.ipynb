```python
def i2(data):
    i = []
    for n in range(len(data)):
        _i = data[n][2]
        i.append(_i)
    return i

i = i2(data)

def a3(data):
    a = []
    for n in range(len(data)):
        _a = data[n][3]
        a.append(_a)
    return a

a = a3(data)

def logxi5(data):
    x = []
    for n in range(len(data)):
        _x = data[n][5]
        x.append(_x)
    return x

logxi = logxi5(data)

def logn6(data):
    y = []
    for n in range(len(data)):
        _y = data[n][6]
        y.append(_y)
    return y
logn = logn6(data)

def afe7(data):
    f = []
    for n in range(len(data)):
        _f = data[n][7]
        f.append(_f)
    return f

afe = afe7(data)
```

```python
chain3 = np.vstack([a,i,logxi,afe])
d = chain3.T
label = [r'$a*$',r'$i$',r'$log\xi$',r'$Afe$']
q = [0.997,0.954,0.683]
qqq = np.percentile(d, [5, 50, 95], axis=0)
print(qqq)
# quantiles1 = np.percentile(a, [5, 95], axis=0)
# print(quantiles1)
# quantiles2 = np.percentile(i, [5, 95], axis=0)
# print(quantiles2)
# quantiles3 = np.percentile(logxi, [5, 95], axis=0)
# print(quantiles3)
# quantiles4 = np.percentile(afe, [5, 95], axis=0)
# print(quantiles4)
```
